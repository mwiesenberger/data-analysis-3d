```python
fig,ax=plt.subplots(1,1,figsize=(10,3.7),dpi= 120, facecolor='w', edgecolor='k')
plt.plot( time*timescale, timeseries_norm[:,0,12])
plt.ylabel( r"$\tilde n$")
plt.xlabel( "time / ms")
plt.savefig('blob-statistics/timeseries.png',bbox_inches='tight')
plt.show()

plt.title('Mean')
plt.plot( xnorm, mean)
plt.xlabel(r'$x-x_{sep}$ / mm')
plt.ylabel( r'$\langle n \rangle$')
plt.savefig( 'blob-statistics/mean.png',bbox_inches='tight')
plt.show()

plt.title('Relative fluctuations')
plt.plot( xnorm, rms/mean)
plt.xlabel(r'$x-x_{sep}$ / mm')
plt.ylabel( r'$n_{rms}/\langle n \rangle$')
plt.savefig( 'blob-statistics/fluctuations.png',bbox_inches='tight')
plt.show()

plt.title('rms')
plt.plot( xnorm, rms)
plt.xlabel(r'$x-x_{sep}$ / mm')
plt.ylabel( r'$n_{rms}$')
plt.savefig( 'blob-statistics/rms.png',bbox_inches='tight')
plt.show()

plt.title('Skewness')
plt.plot( xnorm, skewness)
plt.xlabel(r'$x-x_{sep}$ / mm')
plt.ylabel( r'$S_n$')
plt.savefig( 'blob-statistics/skewness.png',bbox_inches='tight')
plt.show()

plt.title('Flatness')
plt.plot( xnorm, flatness)
plt.xlabel(r'$x-x_{sep}$ / mm')
plt.ylabel( r'$F_n$')
plt.savefig( 'blob-statistics/flatness.png',bbox_inches='tight')
plt.show()

plt.title('Flatness vs skewness')
plt.plot( skewness, flatness, marker = 'x')
plt.plot( skewness, 3 + 3*skewness**2/2)
plt.xlabel(r'$S_n$')
plt.ylabel( r'$F_n$')
plt.savefig( 'blob-statistics/flat-vs-skew.png',bbox_inches='tight')
plt.show()
```

```python
import scipy.special as sps  
import scipy.stats as stats
idx = 10
counts,bins = np.histogram( timeseries_norm[:,:,idx], bins='auto',density=True)
# Gamma distribution function
gamma = mean[idx]**2/rms[idx]**2
print( "Shape is ", gamma)

#https://numpy.org/doc/stable/reference/random/generated/numpy.random.gamma.html
# y = bins**(gamma-1)*(np.exp(-bins/scale)/  (sps.gamma(gamma)*scale**gamma))
##### RENORMALIZE according to Garcia 2016
y = gamma**(0.5+gamma-1)*np.exp(-gamma)/sps.gamma(gamma)*(bins/np.sqrt(gamma) +1)**(gamma - 1)*np.exp(-np.sqrt(gamma)*bins)#/rms[idx]
# Not sure why we should not divide by rms as proposed in the paper
print("x = ", xnorm[idx])

plt.stairs(counts,bins)
# kde = stats.gaussian_kde(timeseries_norm[:,:,idx].flatten())
# plt.plot( bins, kde(bins))
plt.plot( bins, y )
plt.title( 'PDF')
plt.yscale("log")
plt.xlabel( r'$\tilde n$')
plt.savefig('blob-statistics/pdf.png',bbox_inches='tight')
plt.show()
```

```python
# power spectral density
import scipy.signal as signal
frequencies, psd = signal.welch(timeseries_norm[:,0,idx],fs=200e3)

taud = 5e-6
print( "Tau_d is ", taud)
y = 20*rms[idx]**2* 2*taud/(1+taud**2*(2*np.pi*frequencies)**2)
# Plot the PSD
plt.plot(frequencies, psd)
plt.plot(frequencies,y)

plt.xscale('log')
plt.yscale('log')
plt.xlabel('Frequency')
plt.ylabel('Power Spectral Density [V**2/Hz]')
plt.savefig( 'blob-statistics/psd.png', bbox_inches='tight')
plt.show()
```
